```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances

penguins = sns.load_dataset("penguins").dropna()
```

#### Use the pairplot() function on the penguins dataset. Do you visually notice any clusters? How many clusters do you think there are?

```python
sns.pairplot(penguins)
```

Ik zie over het algemeen twee tot drie verschillende clusters. Zo zie ik er een die door het midden loopt van links onder naar rechts boven en ik zie ook twee aparte clusters onder en boven die ook naar boven toe lopen.

#### Use the KMeans algorithm to create a cluster model. Apply this model to the dataset to create an extra column 'cluster' just like we did for the iris dataset above.

#### Use the pairplot(hue='cluster') function to observe how the model has clustered the data.

```python
features = ['bill_length_mm','bill_depth_mm','flipper_length_mm', 'body_mass_g']
km = KMeans(n_clusters=3, random_state=43).fit(penguins[features])
penguins['cluster'] = km.predict(penguins[features])
sns.pairplot(penguins, hue="cluster")
plt.show()
```

Hier is dus te zien dat er drie verschillende clusters zijn voor de data. Ik keek eerst verkeerd naar hoe de clusters werkte maar begrijp nu dat de clusters anders werken. Ik had wel gelijk dat het er 3 zijn.
