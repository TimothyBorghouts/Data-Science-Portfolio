Perform a bivariate analysis on at least 1 combination of 2 columns with categorical data in the dataset that you chose in portfolio assignment 4.
- Do you expect their to be a correlation between the two columns?
- Create a contingency table. Do you observe different ratios between categories here?
- Create a bar plot for this contingency table. Do you observe different ratios between categories here?
- Do a chi-squared test. What does the result say? What's the chance of there being a correlation between the two columns?

```python
import matplotlib.pyplot as plt
import pandas as pd
pokémon = pd.read_csv('pokemon.csv', sep=',')
```

```python
def create_contingency_table(dataset, column1, column2):
    return dataset.groupby([column1, column2]).size().unstack(column1, fill_value=0)

from scipy.stats import chi2_contingency
def check_correlation(dataset, column1, column2):
    contingency_table = create_contingency_table(dataset, column1, column2)
    chi2 = chi2_contingency(contingency_table)
    p_value = chi2[1]
    odds_of_correlation = 1 - p_value
    print(f"The odds of a correlation between {column1} and {column2} is {odds_of_correlation * 100}% (Based on a p value of {p_value}).")
    print("This percentage needs to be at least 95% for a significant correlation.")
```

## Legendary vs Type
### Expectation
Ik verwacht niet dat er teveel correlatie is tussen of een pokémon legendary is of welk type dat de pokémon heeft.

### Contingency table

```python
pokémonContingencyTable = create_contingency_table(pokémon, 'Legendary','Type 1')
print(pokémonContingencyTable)
```

In de contingency tabel is duidelijk te zien dat de meeste legendaries duidelijk van het draken of fisieke type zijn. Ook zie je types waarvan er helemaal geen legendaries zijn zoals bij de insect, vecht, en vergif typen.

### Barplot

```python
plt.rcParams["figure.figsize"] = (12,10)
pokémonContingencyTable.plot(kind='bar')
```

In deze barplot is goed te zien hoeveel er van elk type pokémon zijn en ook hoeveel legendaries er van elk type zijn. Je kunt ook goed de verhoudingen zien. Zo is duidelijk te zien dat er ongeveer water en ground legendaries zijn maar er zijn veel meer water pokémon dus relatief gezien zijn er van de ground types meer legendaries dan de water type.

### Chi-squared test

```python
check_correlation(pokémon, 'Legendary','Type 1')
```

Volgens de chi-squared test is er een hele grote correlatie tussen welk type een pokémon heeft en of die legendary is of niet. Er is volgens de chi-squared test een 99,9 correlatie. Ik heb wel ligte twijfels over deze uitkomst.

### Conclusion
Er is een verband tussen het type en of een pokémon legendary is of niet. Zo is in eerst instantie al goed te zien dat draken en fisieke typen meer legendary pokémons heeft. Ook is snel te zien dat insect, vecht en vergif pokémons geen legendarische pokémons heeft. En sommige typen hebben relatief gezien meer legendaries dan andere. Zo zijn er evenveel vliegende legendaries als niet legendaries. Er is een duidelijk verband.
