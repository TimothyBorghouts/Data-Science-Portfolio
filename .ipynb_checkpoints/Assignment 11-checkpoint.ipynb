```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.head()
```

Als eerst heb ik gekeken naar de verschillende vormen van de pinguins. Hieronder is te zien dat de  Gentoo pinguins meer body mass en flipper lengte hebben. De Adelie pinguins hebben minder bill_length. Tussen de verschillende soorten pinguins zijn er dus wel duidelijk verschillen te zien.

```python
penguins.groupby(by="species").mean()
```

## (1/3) Categorie: Species
Hieronder staan alle barplots voor alle numerieke waardes en de soorten pinguins.

Er is duidelijk te zien dat er een duidelijk verschil is tussen de mediaan waarde voor de verschillende soorten.

Zo is bij de Adelie duidelijk te zien dat die een stuk lagere bill_length heeft en dat de Gentoo een stuk lagere bill_depth heeft. De gento heeft voor de rest ook meer body mass en langere flippers.

```python
sns.barplot(x="species", y="bill_length_mm", data=penguins)
```

```python
sns.barplot(x="species", y="bill_depth_mm", data=penguins)
```

```python
sns.barplot(x="species", y="flipper_length_mm", data=penguins)
```

```python
sns.barplot(x="species", y="body_mass_g", data=penguins)
```

## (2/3) Categorie: Island
Hieronder staan alle barplots voor alle numerieke waardes en de verschillende eiland met pinguins.

Er is duidelijk te zien dat er een duidelijk verschil is tussen de mediaan waarde voor de verschillende soorten.

De pinguins in Torgersen hebben minder bill_length. Op het Biscoe eiland hebben de pinguins minder bill_depth dan de rest. En de flipper lengte en de body mass is allebei groter op het Biscoe eiland.

```python
sns.barplot(x="island", y="bill_length_mm", data=penguins)
```
